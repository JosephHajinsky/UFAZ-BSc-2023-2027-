## **Exercise 1:* Coin Flips*

**Objective:** Analyze the outcomes of simulated coin flips to understand various probability

### **Tasks:**

1. Simulate coin flips: Write a script or program to simulate flipping a fair coin N times
(e.g., N = 1000). Store the results in a data structure (e.g., list, array).


*Solution:*
    In this task, we simulated flipping a fair coin N times, where heads were represented by 1 and tails by 0. This simulation provided us with a dataset of coin flip outcomes, allowing us to analyze various probability concepts.

```python
import numpy as np
import matplotlib.pyplot as plt
from collections import Counter

def simulate_coin_flips(N):
    # Simulate coin flips with 0 as tails and 1 as heads
    results = np.random.choice([0, 1], size=N)
    return results

# Exercise Parameters
N = 1000

# Task 1: Simulate coin flips
results = simulate_coin_flips(N)
```

2. Calculate basic probabilities: Calculate the probabilities of getting heads, tails, and
sequences like HH, HT, TH, TT (for N = 4 flips).

*Solution:*
    We calculated the probabilities of getting heads, tails, and specific sequences (HH, HT, TH, TT) for a small number of flips (N=4). This allowed us to understand the fundamental probabilities associated with a sequence of coin flips.

```python
def calculate_basic_probabilities(results):
    # Calculate basic probabilities
    heads_prob = np.mean(results == 1)
    tails_prob = 1 - heads_prob  # Since there are only two outcomes (heads or tails)

    hh_prob = np.mean((results[:-1] == 1) & (results[1:] == 1))
    ht_prob = np.mean((results[:-1] == 1) & (results[1:] == 0))
    th_prob = np.mean((results[:-1] == 0) & (results[1:] == 1))
    tt_prob = np.mean((results[:-1] == 0) & (results[1:] == 0))

    return heads_prob, tails_prob, hh_prob, ht_prob, th_prob, tt_prob

# Task 2: Calculate basic probabilities
heads_prob, tails_prob, hh_prob, ht_prob, th_prob, tt_prob = calculate_basic_probabilities(results)
```

3. Plot histograms and probability mass functions (PMFs): Create histograms and PMFs for
the number of heads obtained in N flips. Analyze their shape and interpret the results.

*Solution:*
    By creating histograms and PMFs for the number of heads obtained in N flips, we gained insights into the distribution of outcomes. The shapes of these distributions helped us interpret the likelihood of different numbers of heads occurring.

```python
def plot_histogram_and_pmf(results):
    # Plot histogram and PMF
    counts = Counter(results)
    plt.bar(counts.keys(), counts.values(), color=['blue', 'red'])
    plt.title('Histogram of Coin Flips')
    plt.xlabel('Outcome')
    plt.ylabel('Frequency')
    plt.show()

    pmf = np.array(list(counts.values())) / len(results)
    plt.bar(counts.keys(), pmf, color=['blue', 'red'])
    plt.title('Probability Mass Function (PMF) of Coin Flips')
    plt.xlabel('Outcome')
    plt.ylabel('Probability')
    plt.show()

# Task 3: Plot histograms and PMFs
plot_histogram_and_pmf(results)
```

4. Expected value and variance: Calculate the expected value and variance of the number
of heads obtained in N flips. Explain what these values tell you about the distribution of
outcomes.

*Solution:*
    Calculating the expected value and variance of the number of heads in N flips provided us with statistical measures that describe the central tendency and spread of the distribution. These values are crucial for understanding the overall behavior of the random variable.

```python
def calculate_expected_value_and_variance(results):
    # Calculate expected value and variance
    num_heads = np.sum(results == 1)
    expected_value = np.mean(num_heads)
    variance = np.var(num_heads)

    
    return expected_value, variance

# Task 4: Expected value and variance
expected_value, variance = calculate_expected_value_and_variance(results)
print(f"Expected Value: {expected_value}, Variance: {variance}")
```

5. Conditional probabilities: Simulate flipping two coins simultaneously and calculate the
conditional probability of getting tails on the second coin given heads on the first.

*Solution:*
    Simulating the flipping of two coins simultaneously allowed us to calculate the conditional probability of getting tails on the second coin given heads on the first. This helped us understand how the outcome of one coin flip can influence the outcome of another.

```python
def conditional_probability(results):
    # Simulate flipping two coins simultaneously
    first_coin_heads = results[:-1] == 1
    second_coin_tails_given_heads = results[1:][first_coin_heads] == 0
    conditional_prob = np.mean(second_coin_tails_given_heads)

    return conditional_prob
# Task 5: Conditional probabilities

conditional_prob = conditional_probability(results)
print(f"Conditional Probability (Tails on 2nd coin | Heads on 1st coin): {conditional_prob}")
```

6. Law of large numbers: Increase N significantly (e.g., N = 100000) and repeat calculations
from points 2-4. Observe how the results converge towards theoretical probabilities as N
increases.

*Solution:*
    By increasing N significantly (N=100000), we observed how the results converged towards theoretical probabilities. This demonstrated the Law of Large Numbers in action, highlighting the tendency for observed probabilities to approach theoretical probabilities as the sample size increases.

```python
N_large = 100000
results_large = simulate_coin_flips(N_large)

# Repeating tasks 2-4 with larger N
heads_prob_large, tails_prob_large, _, _, _, _ = calculate_basic_probabilities(results_large)
expected_value_large, variance_large = calculate_expected_value_and_variance(results_large)

# Display results for larger N
print("\nResults for Larger N (N = 100000):")
print(f"Heads Probability: {heads_prob_large}")
print(f"Tails Probability: {tails_prob_large}")
print(f"Expected Value: {expected_value_large}, Variance: {variance_large}")
plot_histogram_and_pmf(results_large)
```
